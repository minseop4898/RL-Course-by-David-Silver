```python
import numpy as np
```

```python
def environment(tup, action):
    if (tup[0]==0 and action==1)or(tup[0]==3 and action==3)or(tup[1]==0 and action==0)or(tup[1]==3 and action==2):
        return tup
    if action==0:
        return (tup[0], tup[1]-1)
    elif action==1:
        return (tup[0]-1, tup[1])
    elif action==2:
        return (tup[0], tup[1]+1)
    elif action==3:
        return (tup[0]+1, tup[1])
    return
```

```python
v_k = np.zeros((4,4))
v_k1 = np.zeros((4,4))

for k in range(300):
    for i in range(4):
        for j in range(4):
            if (i==0 and j==0) or (i==3 and j==3): continue
            else:
                sum_ = 0.0
                for a in range(4):
                    tup = environment((i,j),a)
                    sum_ += (-1 + v_k[tup])/4.0
                v_k1[i,j] = sum_
    if k>10:
        v_k = v_k1.copy()
        continue
    print('k = %d'%k)
    print(v_k)
    v_k = v_k1.copy()
print('k = inf')
print(v_k)
```
